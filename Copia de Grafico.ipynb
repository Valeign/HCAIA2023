```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lista=[]
lista2 =[]
flujos = []

for i in range(10, 26):
    f = 100**((21-i)/5)
    logaritmo = np.log(f)
    flujos.append(f)
    lista.append(logaritmo)
    lista2.append(i)
```

```python
# Crear el gráfico
fig, ax = plt.subplots()

# Graficar los datos
ax.plot(lista, lista2)

# Invertir el eje y
ax.set_xlabel('Log(cuentas totales)')
ax.set_ylabel('Magnitud')
ax.set_title('Cuentas totales')
# Mostrar el gráfico
plt.savefig('Grafico1')
```

```python
#Calculo de la SNR para cada magnitud:

snr = []

for j in flujos:
    SNR_arriba = (j*2)*300
    SNR_abajo = (j*2+2*3.14*5*1.3*2)*300 + 4524
    SNR = SNR_arriba/(SNR_abajo)**(0.5)
    print(SNR)
    logaritmo = np.log(SNR)
    snr.append(logaritmo)
    #print(logaritmo)
```

```python
# Crear el gráfico
fig, ax = plt.subplots()

# Graficar los datos
ax.plot(snr, lista2)

# Invertir el eje y

ax.set_xlabel('log(SNR)')
ax.set_ylabel('Magnitud')
ax.set_title('Señal-ruido')
# Mostrar el gráfico
plt.savefig('Grafico2')
```
