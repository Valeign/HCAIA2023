```python
#Leer los datos

import pandas as pd
from scipy.stats import linregress
import math
from scipy.interpolate import interp1d


lc = pd.read_csv('AP31396872.csv', usecols= ['hjd','mag', 'mag_err'])
```

```python
#Funciones:

def promedio():
    suma = 0
    for i in lc['mag']:
        suma = suma + i
    return suma/len(lc['mag'])


def mediana():
    #ordenar los datos:
    lista = sorted(lc['mag'])
    cant = len(lista)
    
    #Condiciones
    if cant%2==0:
        mediana = (lista[cant // 2 - 1] + lista[cant // 2]) / 2
    
    else:
        mediana = lista[cant // 2]
    
    return mediana 


def rango():
    lista = sorted(lc['mag'])
    minimo = lista[0]
    maximo = lista[-1]
    
    return maximo-minimo

def percentiles():
    lista = sorted(lc['mag'])
    
    p20_index =  int(0.2 * (len(lista) + 1)) - 1
    p20 = lista[p20_index]
    
    p35_index = int(0.35* (len(lista) + 1)) - 1
    p35 = lista[p35_index]
    
    p50_index = int(0.5* (len(lista) + 1)) - 1
    p50 = lista[p50_index]
    
    p65_index = int(0.65* (len(lista) + 1)) - 1
    p65 = lista[p65_index]
    
    p80_index = int(0.8* (len(lista) + 1)) - 1
    p80 = lista[p80_index]
    
    return p20, p35, p50, p65, p80

mediana = mediana()
promedio = promedio()
    
def amplitud(mediana):
    lista = sorted(lc['mag'])
    amplitud = lista[-1] - mediana
    return amplitud


def varianza(promedio):
    suma = 0
    for i in lc['mag']:
        cantidad = (i -promedio)**2
        suma += cantidad
    return suma/len(lc['mag'])

def kurtosis(promedio):
    suma = 0
    suma2 = 0
    
    N =  len(lc['mag'])
    
    for i in lc['mag']:
        cantidad = (i -promedio)**4
        suma += cantidad
        
        cantidad2 = (i -promedio)**2
        suma2 += cantidad2
        
        
    kurtosis = ((1/N)*suma)/ ((suma2/N)**2) -3
        
    return kurtosis


def autocorrelacion(promedio, lag=1):
    lista = lc['mag']
    n = len(lista)
    
    suma = 0
    suma2 = 0
    
    for i in range(n - lag):
        cantidad = (lista[i] - promedio) * (lista[i + lag] - promedio)
        suma += cantidad
        
        cantidad2 = (lista[i] - promedio) ** 2
        suma2 += cantidad2
    
    autocorrelacion = suma / suma2
        
    return autocorrelacion

def asimetria(promedio):
    
    lista = lc['mag']
    n = len(lista)
    
    suma = 0
    suma2 = 0
    
    for i in range(0, n-1):
        cantidad = (lista[i] - promedio)**3
        suma += cantidad
        
        cantidad2 = (lista[i] - promedio) ** 2
        suma2 += cantidad2
    
    asimetria = (suma/n)/((suma2/n)**(3/2))
        
    return asimetria
    
    

varianza = varianza(promedio)
    
def desviacion_est(varianza):
    desviacion = varianza**(1/2)
    return desviacion 

def pendiente_al():
    x = lc['mag']
    y = lc['hjd']
    
    m, n, r_value, p_value, std_err = linregress(x, y)
    
    return m

from scipy.fft import fft, fftfreq
import numpy as np

```

```python
from scipy.fft import fft, fftfreq


def fourier():
    """
    Calcula los componentes de la serie de Fourier de una curva de luz.

    Parámetros:
    - tiempo: np.array
        Array que representa el tiempo.
    - intensidad: np.array
        Array que representa la intensidad de la curva de luz.
    - num_armonicos: int, opcional
        Número máximo de armónicos a incluir.

    Retorna:
    - armonicos: dict
        Diccionario que contiene las frecuencias, amplitudes y fases de los armónicos.
    """
    tiempo = lc['hjd']
    intensidad = lc['mag']
    
    num_armonicos = 4
    
    
    
    # Frecuencia de muestreo
    fs = 1 / (tiempo[1] - tiempo[0])

    # Calcula la FFT y las frecuencias asociadas
    fft_result = fft(intensidad)
    frequencies = fftfreq(len(intensidad), 1/fs)

    # Encuentra las amplitudes y fases de los componentes en la FFT
    amplitudes = np.abs(fft_result) / len(intensidad)
    phases = np.angle(fft_result)

    # Filtra solo las frecuencias positivas (la mitad del espectro)
    positive_frequencies = frequencies[:len(frequencies)//2]
    positive_amplitudes = 2 * amplitudes[:len(amplitudes)//2]  # Se multiplica por 2 para corregir la amplitud

    # Selecciona los primeros 'num_armonicos' armónicos
    armonicos = {
        'frecuencias': positive_frequencies[:num_armonicos],
        'amplitudes': positive_amplitudes[:num_armonicos],
        'fases': phases[:num_armonicos]
    }

    return armonicos

print(fourier())
```
